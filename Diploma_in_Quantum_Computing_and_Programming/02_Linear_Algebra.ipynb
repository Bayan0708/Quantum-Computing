###**Definición de matriz**

Una matriz  $A_{n\times m}$, de tamaño $n \times m$ es un arreglo rectangular de $n \times  m$ números reales (o complejos) ordenados en n filas (renglones) horizontales y m columnas (verticales).

Otra manera de pensarla es como un conjunto ordenado en una estructura de filas y columnas, los elementos de este conjunto pueden ser objetos matemáticos de diferentes tipos, sin embargo, trabajaremos exclusivamente con matrices formadas por números reales (aunque puede ser cualquier elemento de un <a href=https://es.wikipedia.org/wiki/Espacio_vectorial>*Espacio Vectorial*</a>, para saber mas de este tema visita nuestro material de <a href=https://colab.research.google.com/github/jugernaut/Prometeo/blob/desarrollo/02_AlgebraLineal/05_Espacios_Vectoriales/05_Espacios_Vectoriales.ipynb>*Espacios Vectoriales*</a>).

Cada uno de los números  que conforman la matriz se denomina elemento y se denotan por $(a_{ij})$ donde $i$ representa el número de fila y $j$ el número de la columna en que se encuentra el elemento de la matriz $A$. El número de filas y columnas de una matriz respresenta la  dimensión de una matriz, por ejemplo una matriz A de $n$ filas y $m$ columnas podemos denotarla como $A_{n\times m}$ ,  es decir:

$$ \begin{equation} A_{n\times m} =
\begin{pmatrix}
a_{11} & a_{21} & \cdots & a_{1m}\\
a_{21} & a_{22} & \cdots & a_{2m}\\
\vdots & \vdots & \ddots & \vdots\\
a_{n1} & a_{n2} & \cdots & a_{nm}
\end{pmatrix}
\end{equation} $$


A continuación se presentará un ejemplo en el que identificaremos la dimensión de la matriz, asi como sus elementos.

### **Suma  y Resta de matrices**

Para llevar a cabo estas operaciones es importante verificar que las matrices con las que se trabajará tengan el mismo orden, también es importante saber que se obtendrá como resultado una matriz con exactamente ese orden al realizar las operaciones.

La suma o resta se lleva a cabo elemento por elemento, por ejemplo, sean $\textbf{A}$ y $\textbf{B}$ dos matrices que deseamos sumar, entonces se obtendrá una nueva matriz $\textbf{C}$ cuyos elementos serán: $c_{ij}=a_{ij}\pm b_{ij}$, para hacer la resta solo le aplica la operación correspondiente.

De manera gráfica, podemos verlo de la siguiente manera para la suma:

$$ \begin{equation} A_{n\times m} =
\begin{pmatrix}
a_{11} & a_{12} & \cdots & a_{1m}\\
a_{21} & a_{22} & \cdots & a_{2m}\\
\vdots & \vdots & \ddots & \vdots\\
a_{n1} & a_{n2} & \cdots & a_{nm}
\end{pmatrix}
\end{equation} $$


$$ \begin{equation} B_{n\times m} =
\begin{pmatrix}
b_{11} & b_{12} & \cdots & b_{1m}\\
b_{21} & b_{22} & \cdots & b_{2m}\\
\vdots & \vdots & \ddots & \vdots\\
b_{n1} & b_{n2} & \cdots & b_{nm}
\end{pmatrix}
\end{equation}$$


$$ \begin{equation} C=A\pm B =
\begin{pmatrix}
a_{11}\pm b_{11} & a_{12}\pm b_{12} & \cdots & a_{1m}\pm b_{1m}\\
a_{21}\pm b_{21} & a_{22}\pm b_{22} & \cdots & a_{2m}\pm b_{2m}\\
\vdots & \vdots & \ddots & \vdots\\
a_{n1}\pm b_{n1} & a_{n2}\pm b_{n2} & \cdots & a_{nm}\pm b_{nm}
\end{pmatrix}
\end{equation}$$

Cabe destacar que en esta operación con matrices se mantiene la $\textbf{asociatividad}$, $\textbf{conmutatividad}$, y el elemento $\textbf{neutro}$ que seria la matriz cuyas entradas son todas cero y por último el elemento opuesto el cuál si tenemos la matriz $\textbf{A}$, el elemento opuesto  sería $\textbf{-A}$  que es una matriz cuyos elementos son los opuestos a $\textbf{A}$.

### **Escalar por una matriz**

En este caso lo que se hace es multiplicar un número escalar por la matriz, es decir $\kappa A_{n\times m}$ y el resultado será una matriz del mismo orden pero con cada uno de sus elementos múltiplicado por el escalar,es decir:


  $$\begin{equation} \kappa A_{n\times m}=
\begin{pmatrix}
\kappa a_{11} & \kappa a_{12} & \cdots & \kappa a_{1m}\\
\kappa a_{21} & \kappa a_{12} & \cdots & \kappa a_{2m}\\
\vdots & \vdots & \ddots & \vdots\\
\kappa a_{n1} & \kappa a_{n2} & \cdots & \kappa a_{nm}
\end{pmatrix}
\end{equation}$$

Cabe mencionar que esta operación es $\textbf{distributiva}$ respecto a la suma de escalares y a la suma de matrices, además es conmutativa y asociativa respecto al producto punto de escalares.

### **Producto de un vector por una matriz**

Si tenemos una matriz $A_{n\times m}$ y queremos obtener el producto de esta matriz con un vector, lo primero que se debe verificar es que el número de columnas de la matriz coincida con el número de filas del vector, el resultado de esto nos dara una nueva matriz $C_{n\times 1}$.

Los elementos que conformarán esta nueva matriz serán el resultado de sumar los productos de los elementos de la fila $i$ por el vector, es decir:

Sea la matriz:

 $$ \begin{equation}  A_{n\times m} =
\begin{pmatrix}
a_{11} & a_{12} & \cdots & a_{1m}\\
a_{21} & a_{12} & \cdots & a_{2m}\\
\vdots & \vdots & \ddots & \vdots\\
a_{n1} & a_{n2} & \cdots & a_{nm}
\end{pmatrix}
\end{equation}$$

Definiremos el vector:

  $$\begin{equation}  V_{m\times 1} =
\begin{pmatrix}
v_{11} \\
v_{21} \\
\vdots \\
v_{m1}
\end{pmatrix}
\end{equation}$$


Entonces el producto de estos será la nueva matriz $C_{n\times 1}$ cuyas entradas serán:


$$
\begin{equation}
  C_{i1} = \sum_{k} (a_{ik}*v_{k1})
\end{equation}$$

### **Producto de Matrices**

Si tenemos una matriz $A_{n\times m}$ y queremos obtener el producto de esta matriz con una $B_{m\times l}$, lo primero que se debe verificar es que el número de columnas de la matriz $A$ coincida con el número de filas de la matriz $B$, el resultado de esto nos dara una nueva matriz $C_{n\times l}$.

Los elementos que conformarán esta nueva matriz serán el resultado de sumar los productos de los elementos de la fila $i$ por la columna $j$, es decir:

### **Matriz Inversa**

Una matriz cuadrada $A_{n\times n}$ tiene inversa denotada como $A^{-1}$ si se cuemple que:

$$AA^{-1}=A^{-1}A=I$$

De donde $I$ es la matriz identidad de tamaño $n$ la cuál tiene unos en su diagonal y ceros en todas las demas entradas.

Las matrices que tienen inversa son llamadas matrices invertibles o regulares.

La matriz inversa debe cumplir:
 * $(A^{-1})^{-1}$ = $A$
 * $(A B)^{-1}$ = $B^{-1}$$A^{-1}$
 * $(\alpha A)^{-1}$=$\frac{1}{\alpha}A^{-1}$   con $\alpha\neq 0$

### **Producto tensorial**

En algebra lineal, $\textbf{el producto tensorial}$, denotado por $\otimes$, se puede aplicar en diversos contextos a $\textbf{vectores, matrices, tensores y espacios vectoriales}$. En cada caso, el significado del símbolo es el mismo: la operación bilineal más general.

La forma más clara de mostrar cómo funciona el producto tensorial es con un ejemplo:

$$
A\otimes B =\begin{pmatrix} a \\ b \end{pmatrix} \otimes \begin{pmatrix} c \\ d \end{pmatrix} =  
\begin{pmatrix} a \begin{pmatrix} c \\ d \end{pmatrix} \\ b \begin{pmatrix}  c \\ d \end{pmatrix} \end{pmatrix} =
\begin{pmatrix} ac \\ ad \\ bc \\ bd \end{pmatrix}
$$

Esto se generaliza a dimensiones superiores, por ejemplo:

$$
\begin{pmatrix} a_1 & b_1 \\ c_1 & d_1 \end{pmatrix} \otimes \begin{pmatrix} a_2 & b_2 \\ c_2 & d_2 \end{pmatrix} =  
\begin{pmatrix} a_1 \begin{pmatrix} a_2 & b_2 \\ c_2 & d_2 \end{pmatrix} & b_1 \begin{pmatrix} a_2 & b_2 \\ c_2 & d_2 \end{pmatrix} \\ c_1 \begin{pmatrix} a_2 & b_2 \\ c_2 & d_2 \end{pmatrix} & d_2 \begin{pmatrix} a_2 & b_2 \\ c_2 & d_2 \end{pmatrix} \end{pmatrix} =
\begin{pmatrix}
a_1 a_2 & a_1 b_2 & b_1 a_2 & b_1 b_2 \\
a_1 c_2 & a_1 d_2 & b_1 c_2 & b_1 d_2 \\
c_1 a_2 & c_1 b_2 & d_1 a_2 & d_1 b_2 \\
c_1 c_2 & c_1 d_2 & d_1 c_2 & d_1 d_2 \\
\end{pmatrix}
$$

### **Producto escalar**

El producto escalar, también conocido como producto punto, es una operación matemática que toma dos $\textbf{vectores}$ y devuelve un $\textbf{escalar}$ (un número real) como resultado.

$$\begin{split}
A =& \begin{pmatrix} a_1 \\ a_2 \end{pmatrix};\qquad B = \begin{pmatrix} b_1 \\ b_2 \end{pmatrix} \\\\
A \cdot B =&~ a_1 \times b_1 + a_2 \times b_2
\end{split}$$

o sea, la suma de los productos componente por componente.

El producto escalar se utiliza para definir la _norma_, o longitud, de un vector:

$$\rightarrow |A| = \sqrt{A \cdot A} = \sqrt{a_{1}^2 + a_{2}^2}$$

### **Representación matricial de los qubits**

Los elementos $\{\vert 0\rangle, \vert 1\rangle\}$, son los vectores:

# **Representación matricial de los operadores**

Un $\textbf{operador cuántico}$ que actúa sobre $n$ qubits, esta representado mediante una matriz unitaria de $2^n\times 2^n$. Los vectores que representan estados cuánticos de dos niveles (qubits) sobre los que actúan los operadores.

La acción de un operador en un estado cuántico se aplica multiplicando el vector $\vert \psi_{1}\rangle$, por la matriz $U$ que representa el operador. El resultado es un nuevo vector $\vert \psi_{2}\rangle$:

$$U\vert\psi_{1} \rangle=\vert\psi_{2} \rangle$$

Cuando tenemos un qubit, el vector que lo representa tiene dos elementos y un operador que actua sobre un qubit es una matriz de $2\times 2$:

$$U\vert \psi_{1}\rangle =
\begin{bmatrix}
    u_{11} & u_{12} \\
    u_{21} & u_{22}
\end{bmatrix}
\begin{bmatrix}
    \alpha \\
    \beta
\end{bmatrix}
=
\begin{bmatrix}
    \alpha \, u_{11} + \beta \, u_{12} \\
    \alpha \, u_{21} + \beta \, u_{22}
\end{bmatrix}
=
\begin{bmatrix}
    \alpha' \\
    \beta'
\end{bmatrix}
= \vert \psi_{2}\rangle
$$

### **Matriz de Hadamard**

En matemáticas, una $\textbf{matriz de Hadamard}$, llamada así por el matemático francés $\textbf{Jacques Hadamard}$, es una matriz cuadrada cuyas entradas son $+1$ o $−1$ y cuyas filas son mutuamente ortogonales.

Su representación matricial es,

$$H =\frac{1}{\sqrt{2}} \begin{bmatrix}1 & 1 \\
 1 & -1\end{bmatrix}$$

$$
H \vert 0\rangle = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix} \begin{bmatrix} 1 \\ 0 \end{bmatrix} = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$

$$
\begin{split}
H |0\rangle =&~ \frac{1}{\sqrt{2}} \left(|0\rangle + |1\rangle \right)\\
H |1\rangle =&~ \frac{1}{\sqrt{2}} \left(|0\rangle - |1\rangle \right)
\end{split}
$$

Nótese que $H$ también es su propio inverso, tal que $HH = I$.

```python
# --- PARTE 1: RESOLUCIÓN DE LA ECUACIÓN DE SCHRÖDINGER INDEPENDIENTE DEL TIEMPO ---

print("--- Resolución de la Ecuación de Schrödinger Independiente del Tiempo ---")

# 1. Definir la matriz Hamiltoniana (H)
# Para un sistema de dos niveles, H es una matriz de 2x2.
# Este es un ejemplo simple, donde los elementos diagonales son las energías de los estados
# "puros" y los elementos fuera de la diagonal representan el acoplamiento entre ellos.
# Los valores están en unidades arbitrarias de energía (e.g., eV, Joules).
# Usaremos valores que no sean triviales para ver una mezcla de estados.

# Ejemplo de Hamiltoniano para un sistema de dos niveles
# (e.g., espín en un campo magnético o interacción entre dos estados):
# H = | E1  V  |
#     | V   E2 |
# Donde E1 y E2 son las energías de los estados base y V es el acoplamiento.
E1 = 1.0  # Energía del estado base 1
E2 = 3.0  # Energía del estado base 2
V = 0.5   # Acoplamiento entre los estados

H = np.array([[E1, V],
              [V, E2]])

print("\nMatriz Hamiltoniana (H):\n", H)

# 2. Encontrar los valores propios (energías) y vectores propios (estados propios)
# Usamos np.linalg.eigh porque H es una matriz Hermitiana (simétrica en este caso).
# Esto nos da los valores propios (energías posibles del sistema) y los vectores propios correspondientes
# (las funciones de onda estacionarias).

eigenvalues, eigenvectors = np.linalg.eigh(H)

print("\nValores propios (Energías del sistema):\n", eigenvalues) # Estas son las energías E del sistema

print("\nVectores propios (Estados estacionarios - normalizados por columna):\n", eigenvectors)
# Cada columna de 'eigenvectors' es un vector propio.
# eigenvectors[:, 0] es el vector propio para eigenvalues[0]
# eigenvectors[:, 1] es el vector propio para eigenvalues[1]

# Interpretación de los vectores propios:
# Si un vector propio es [c1, c2], significa que el estado es c1*|estado_base_1> + c2*|estado_base_2>.
# |c1|^2 es la probabilidad de encontrar el sistema en el estado base 1.
# |c2|^2 es la probabilidad de encontrar el sistema en el estado base 2.

print("\nInterpretación de los Vectores propios:")
for i, energy in enumerate(eigenvalues):
    state_vector = eigenvectors[:, i]
    prob_state1 = np.abs(state_vector[0])**2
    prob_state2 = np.abs(state_vector[1])**2
    print(f"  Energía E{i+1}: {energy:.4f}")
    print(f"    Vector de estado |ψ{i+1}>: {state_vector}")
    print(f"    Probabilidad de estar en el estado 1: {prob_state1:.4f}")
    print(f"    Probabilidad de estar en el estado 2: {prob_state2:.4f}")
    print(f"    (Normalización: {prob_state1 + prob_state2:.4f})") # Debería ser 1
```

### **Ecuacion de Schrödinger dependiente del tiempo**

$$i\hslash\frac{d}{dt}\vert\psi(t)\rangle = \hat{H}\vert\psi(t)\rangle$$

```python
# --- PARTE 2: RESOLUCIÓN DE LA ECUACIÓN DE SCHRÖDINGER DEPENDIENTE DEL TIEMPO ---

print("\n\n--- Resolución de la Ecuación de Schrödinger Dependiente del Tiempo ---")

# La solución general es |ψ(t)> = exp(-iHt/hbar) |ψ(0)>
# Definimos una constante de Planck reducida (hbar).
# Para simplificar, podemos establecer hbar = 1 para obtener unidades de tiempo "naturales".
hbar = 1.0 # Constante de Planck reducida (en unidades arbitrarias)

# 1. Definir un estado inicial |ψ(0)>
# Vamos a suponer que el sistema comienza completamente en el estado base 1.
# En la base [|estado_base_1>, |estado_base_2>], esto es [1, 0].
psi_0 = np.array([1.0, 0.0]) # Estado inicial: completamente en el estado base 1
print(f"\nEstado inicial |ψ(0)>: {psi_0}")

# 2. Definir un rango de tiempo
times = np.linspace(0, 10, 100) # 100 puntos de tiempo desde 0 hasta 10 (unidades de tiempo)

# 3. Calcular la evolución del estado |ψ(t)> en cada instante de tiempo
psi_at_time = []
probabilities_state1 = []
probabilities_state2 = []

for t in times:
    # Calcular el operador de evolución temporal U(t) = exp(-iHt/hbar)
    # scipy.linalg.expm calcula la exponencial de la matriz.
    evolution_operator = expm(-1j * H * t / hbar) # i es 1j en Python

    # Calcular el estado en el tiempo t: |ψ(t)> = U(t) |ψ(0)>
    current_psi = np.dot(evolution_operator, psi_0)
    psi_at_time.append(current_psi)

    # Calcular las probabilidades de encontrar el sistema en cada estado base
    prob1 = np.abs(current_psi[0])**2
    prob2 = np.abs(current_psi[1])**2
    probabilities_state1.append(prob1)
    probabilities_state2.append(prob2)

# 4. Imprimir o graficar los resultados (para visualizar la evolución)
print("\nEvolución de las probabilidades en el tiempo:")
print("Tiempo (t) | P(Estado 1) | P(Estado 2)")
print("-------------------------------------------------")
for i in range(0, len(times), 10): # Imprime solo algunos puntos para no saturar
    t = times[i]
    p1 = probabilities_state1[i]
    p2 = probabilities_state2[i]
    print(f"{t:9.2f} | {p1:16.4f} | {p2:16.4f}")

# Visualización con Matplotlib

plt.figure(figsize=(10, 6))
plt.plot(times, probabilities_state1, label='P(Estado 1) = |<ψ_base_1|ψ(t)>|^2')
plt.plot(times, probabilities_state2, label='P(Estado 2) = |<ψ_base_2|ψ(t)>|^2', linestyle='--')
plt.xlabel('Tiempo (unidades arbitrarias)')
plt.ylabel('Probabilidad')
plt.title('Evolución Temporal de las Probabilidades')
plt.legend()
plt.grid(True)
plt.show()
```
